```python
#importing necessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#extracting cleaned play store data
df=pd.read_csv('../googleplaystore_clean.csv') 
```

```python
#Category wise visualization of applications
plt.figure(figsize = (8,8))
df["Category"].value_counts().plot(kind='pie')
plt.show()
plt.figure(figsize = (10,5))
df["Category"].value_counts().plot(kind='bar')
plt.ylabel("No of Apps")
plt.xlabel("Category of Apps")
plt.grid()
plt.show()
```

```python
#Visualization based on Content Rating
plt.figure(figsize = (7,7))
df["Content_Rating"].value_counts().plot(kind='pie',autopct = "%f%%")
plt.legend()
plt.show()
df["Content_Rating"].value_counts().plot(kind='bar')
plt.ylabel("No of Apps")
plt.xlabel("Types of Content Rating")
plt.grid()
plt.show()
```

```python
#Visualization based on genre
plt.figure(figsize = (20,10))
df["Genres"].value_counts().plot(kind='bar')
plt.ylabel('No of Apps')
plt.xlabel('Type of Genres')
plt.show()
```

```python
#Types of Apps based on price (Free vs Paid)
plt.figure(figsize = (5,5))
df["Type"].value_counts().plot(kind='pie',autopct = "%i%%")
plt.legend()
plt.show()
```

```python
#Applications with maximum size
df_size_max=df[df['Size'] == df['Size'].max()]
sm=df['Size'].max()
print("Maximum Size of Apps in bytes in the data set is", sm)

print("List of those applications is as follows:")
df_size_max.App
```

```python
#Applications with minimum size
df_size_min=df[df['Size'] == df['Size'].min()]
sm=df['Size'].min()
print("Minimum Size of App in bytes in the data set is", sm)

print("List of those application is as follows:")
print(df_size_min.App)
```

```python
#Applications with maximum installs
df_installs_max=df[df['Installs'] == df['Installs'].max()]
sm=df['Installs'].max()
print("Maximum installs of apps in the data set is", sm)

print("List of those applications is as follows:")
df_installs_max.App
```

```python
#Applications with mainimum installs
df_installs_min=df[df['Installs'] == df['Installs'].min()]
sm=df['Installs'].min()
print("Minimum installs of apps in the data set is", sm)

print("List of those applications is as follows:")
df_installs_min.App
```

```python
#Application with most reviews
df[df['Reviews'] == df['Reviews'].max()]
```

```python
#Application with least reviews
df[df['Reviews'] == df['Reviews'].min()]
```

```python
#Visualizing applications with most reviews
sorte = df.sort_values(['Reviews'],ascending = 0 )[:20]
ax = sns.barplot(x = 'Reviews' , y = 'App' , data = sorte )
ax.set_xlabel('Reviews')
ax.set_title("Most Reviewed Apps in Play Store", size = 20)
```

```python
#Visualizing Applications which are most installed in most popular category "Family"
sort_data = df[df['Category'] == 'FAMILY'].sort_values(['Installs'],ascending= 0)[:15]
ax = sns.barplot(x = 'Installs' , y = 'App' , data = sort_data )
ax.set_xlabel('Installs')
ax.set_title("Most Installed Apps in Play Store", size = 20)
```
